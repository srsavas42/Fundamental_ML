```python
import numpy as np
import pandas as pd
import json

# import our final dataset
with open("/content/drive/My Drive/CS449_Project/CS_449_Data/full_dataset_4.csv", 'r') as csv:
  full_dataset = pd.read_csv(csv)

full_dataset
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# take out only data that will be used for the regression, split by stock category (spcindcd)
regr_data = full_dataset[['return', 'Volume', 'op_income_pre_dep', 'op_income_post_dep', 'sales_net', 'current_assets', 'current_liabilities', 'retained_earnings', 'working_capital', 'return group', 'spcindcd']].dropna().groupby("spcindcd")

# select a ticker to be used for an in-depth look later
ticker = "AMZN"
# loop through the data for each stock category
for i in regr_data:
  # identify the inputs and expected outputs
  X = np.array(i[1][['return', 'Volume', 'op_income_pre_dep', 'op_income_post_dep', 'current_assets', 'current_liabilities', 'retained_earnings', 'working_capital']])
  y = np.array(i[1][['return group']])

  # fit the linear regression
  regr = LinearRegression()
  regr.fit(X, y)
  # run a sample plot for the previously defined ticker by grabbing the correct model for the same spcindcd
  if i[1]['spcindcd'][i[1].index[0]] == full_dataset[full_dataset["ticker"] == ticker]['spcindcd'][full_dataset[full_dataset["ticker"] == ticker].index[0]]:
    temp_X = full_dataset[full_dataset["ticker"]==ticker][['return', 'Volume', 'op_income_pre_dep', 'op_income_post_dep', 'current_assets', 'current_liabilities', 'retained_earnings', 'working_capital']].dropna()
    temp_y = full_dataset[full_dataset["ticker"]==ticker]['return group'].dropna().reset_index(drop=True)

    plt.plot(regr.predict(temp_X))
    plt.plot(temp_y)
    plt.title("Linear Regression Classification of AMZN")

    print('Accuracy of LR', regr.score(temp_X, temp_y))
```
